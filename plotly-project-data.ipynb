```python
cols =['District code','State name','District name','Male','Female','Male_Literate','Female_Literate','Households_with_Internet','Housholds_with_Electric_Lighting']
```

```python
census = census[cols]
census.head(2)
```

```python
final_df = latlong.merge(census,left_on='District',right_on='District name').drop(columns=['District name','State name'])
final_df.head(3)
```

```python
final_df['Sex_Ratio'] = (final_df.Female/final_df.Male)*1000
final_df['Litracy_Rate'] = np.round(((final_df.Male_Literate+final_df.Female_Literate)/(final_df.Male+final_df.Female))*100,2)
```

```python
final_df.to_csv('india.csv',index=False)
```
